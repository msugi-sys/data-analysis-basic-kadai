```python
# matplotlibとnumpyをインポート
import matplotlib.pyplot as plt
import numpy as np

# 描画領域を生成
fig, axes = plt.subplots(2, 3, tight_layout=True)

# グラフの生成
# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)
# sin関数 ０から１０までの幅で配列を等間隔に１００個作成する
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0,1].plot(x1, y1, linewidth=2.0)
axes[0,1].set_xlim(0, 10)
axes[0,1].set_xticks(np.arange(0, 11, 1))
axes[0,1].set_ylim(-1, 5)
axes[0,1].set_yticks(np.arange(-1, 6, 1))

# ヒストグラム 平均が15,標準偏差5の標準正規分布から2000個の乱数を生成する
x2 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x2)
# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3, y3)
# 描画
plt.show()
```
